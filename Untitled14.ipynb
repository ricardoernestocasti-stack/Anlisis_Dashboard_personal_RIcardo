```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el CSV
df = pd.read_csv("/content/anime_dataset.csv")

# Convertir el año a número (por si hay texto o valores vacíos)
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df = df.dropna(subset=["year", "score"])

# Agrupar por año y calcular promedio del score
promedio_score = df.groupby("year")["score"].mean()

# Graficar correctamente
plt.plot(promedio_score.index, promedio_score.values,'go--', label="promedio del score por año")
plt.xlabel("Año")
plt.ylabel("Score promedio")
plt.title("Evolución del puntaje promedio de anime por año")
plt.grid(True)
plt.legend()
plt.savefig("score_promedio_de_cada_año.png")
plt.show()

df["year"] = pd.to_numeric(df["year"], errors="coerce")
df = df.dropna(subset=["year", "popularity"])
promedio_popularity = df.groupby("year")["popularity"].mean()

##grafica de promedio de popularidad por año
plt.plot(promedio_popularity.index, promedio_popularity.values, 'k*-.', label="promedio de popularidad por año")
plt.title("el promedio de de popularidad de un anime por año")
plt.xlabel("año")
plt.ylabel("promedio de popularidad")
plt.grid(True)
plt.legend()
plt.savefig('Popularidad_promedio_de_cada_año.png')
plt.show()


#comparando el promedio de la popularidad con la puntuación
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["popularity"] = pd.to_numeric(df["popularity"], errors="coerce")
df["episodes"] = pd.to_numeric(df["episodes"], errors="coerce")
promedio_comparativo = df.groupby('year')[['popularity', 'episodes']].mean()
plt.plot(promedio_comparativo.index, promedio_comparativo["popularity"], 'k*-.', label="Popularidad promedio")
plt.plot(promedio_comparativo.index, promedio_comparativo["episodes"], "cx-.", label="Episodios promedio")
plt.grid(True)
plt.title("comparacion del promedio entre episodios y su popularidad por año")
plt.xlabel("Año")
plt.ylabel("Promedio")
plt.legend()
plt.savefig("promedio entre episodios y su popularidad por año.png")
plt.show()
```
